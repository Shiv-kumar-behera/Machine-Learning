```python
delivery = pd.read_csv("deliveries.csv")
delivery.head(2)
```

```python
runs = delivery.groupby("batsman")
```

```python
runs["batsman_runs"].sum().sort_values(ascending=False).head(10)
```

```python
mask = delivery["batsman_runs"] == 3
new_delivery = delivery[mask]
new_delivery
```

```python
four_runs = new_delivery.groupby("batsman")
```

```python
four_runs["batsman_runs"].count().sort_values(ascending=False).head()
```

```python
vk = delivery[delivery["batsman"] == "V Kohli"]
vk
```

```python
vk_runs = vk.groupby("bowling_team")
vk_runs["batsman_runs"].sum().sort_values(ascending=False).head(3)
```

```python
def get_player_max_runs_against_teams(player):
    player_stats = delivery[delivery["batsman"] == player]
    player_runs = player_stats.groupby("bowling_team")
    return player_runs["batsman_runs"].sum().sort_values(ascending=False).index[0]
```

```python
get_player_max_runs_against_teams("DA Warner")
```

```python
death_over = delivery[delivery["over"] > 15]
death_over
```

```python
all_batsman = death_over.groupby("batsman")["batsman_runs"].count()
x = all_batsman > 200
all_batsman[x]
```

```python
batsman_list = all_batsman[x].index.tolist()
batsman_list
```

```python
final = delivery[delivery["batsman"].isin(batsman_list)]
final
```

```python
runs = final.groupby("batsman")["batsman_runs"].sum()
runs
```

```python
balls = final.groupby("batsman")["batsman_runs"].count()
balls
```

```python
sr = runs / balls * 100
sr.sort_values(ascending=False).index[0]
```
