```python
data = pd.read_csv("/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
data.head()
```

```python
data['review'][0]
```

```python
data.info()
```

```python
data['sentiment'].replace('positive',1,inplace=True)
data['sentiment'].replace('negative',0,inplace=True)
data.head()
```

```python
import re
clean = re.compile("<.*?>")
re.sub(clean, '', data.iloc[2].review)
```

```python
def remove_html(text):
  clean = re.compile("<.*?>")
  return re.sub(clean, '', text)
    
```

```python
data['review'] = data['review'].apply(remove_html)
data.head()
```

```python
def convert_lower(text):
  return text.lower()
```

```python
data['review'] = data['review'].apply(convert_lower)
data.head()
```

```python
def remove_special(text):
  x = ''
  for i in text:
    if i.isalnum():
      x+=i
    else:
      x += ' '
  return x
```

```python
data['review'] = data['review'].apply(remove_special)
data.head()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
  x = []
  for i in text.split():
    if i not in stopwords.words("english"):
      x.append(i)
  y = x[:]
  x.clear()
  return y
```

```python
data['review'] = data['review'].apply(remove_stopwords)
data.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_words(text):
  y = []
  for i in text:
    y.append(ps.stem(i))
  z = y[:]
  y.clear()
  return z
```

```python
data['review'] = data['review'].apply(stem_words)
data.head()
```

```python
def join_words(input_list):
  return " ".join(input_list)
```

```python
data['review'] = data['review'].apply(join_words)
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
x = cv.fit_transform(data['review']).toarray()
```

```python
x.shape
```

```python
y = data.iloc[:,-1].values
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
model1 = GaussianNB()
model2 = MultinomialNB()
model3 = BernoulliNB()
```

```python
model1.fit(x_train, x_test)
model2.fit(x_train, x_test)
model3.fit(x_train, x_test)
```

```python
y_pred1 = mode1.predict(x_test)
y_pred2 = mode2.predict(x_test)
y_pred3 = mode3.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Guassian:", accuracy_score(y_test, y_pred1))
print("Multinomial:", accuracy_score(y_test, y_pred2))
print("Bernoulli:", accuracy_score(y_test, y_pred3))
```
