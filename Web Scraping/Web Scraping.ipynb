```python
data_set = pd.DataFrame()
for j in range(1, 503):
    url = "https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget&page={}".format(j)
    response = requests.get(url, headers=headers, cookies=cookies)
    soup = BeautifulSoup(response.text, 'lxml')
    company = soup.find_all("div", class_ = "companyCardWrapper")
    name = []
    rating = []
    reviews = []
    category = []
    hq = []
    no_of_branches = []
    salary = []
    no_of_interviews = []
    jobs = []
    for i in company:
        name.append(i.find("h2").text.strip())
        try:
            rating.append(i.find("div", class_ = "rating_text").text.strip())
        except:
            rating.append(np.nan)
        try:
            reviews.append(i.find("span", class_ = "companyCardWrapper__companyRatingCount").text.strip().split(")")[0][1:])
        except:
            reviews.append(np.nan)
        x = i.find("span", class_ = "companyCardWrapper__interLinking").text.strip()
        if("|" in x):
            x = x.split("|")
            try:
                category.append(x[0])
            except:
                category.append(np.nan)
            try:
                hq.append(x[1].split("+")[0].strip())
            except:
                hq.append(np.nan)
            try:
                no_of_branches.append(x[1].split("+")[1].split(" ")[0])
            except:
                no_of_branches.append(np.nan)
        else:
            category.append(np.nan)
            try:
                hq.append(x.split("+")[0].strip())
            except:
                hq.append(np.nan)
            try:
                no_of_branches.append(x.split("+")[1].split(" ")[0])
            except:
                no_of_branches.append(np.nan)
        try:
            salary.append(i.find_all("span", class_ = "companyCardWrapper__ActionCount")[1].text.strip())
        except:
            salary.append(np.nan)
        try:
            no_of_interviews.append(i.find_all("span", class_ = "companyCardWrapper__ActionCount")[2].text.strip())
        except:
            no_of_interviews.append(np.nan)
        try:
            jobs.append(i.find_all("span", class_ = "companyCardWrapper__ActionCount")[3].text.strip())
        except:
            jobs.append(np.nan)
    
    d = {"name": name, "rating": rating, "reviews": reviews, "category": category, "hq": hq, "no_of_branches": no_of_branches, "salary": salary, "no_of_interviews": no_of_interviews, "jobs": jobs}
    df = pd.DataFrame(d)
    data_set = pd.concat([data_set, df], ignore_index=True)
data_set
```
